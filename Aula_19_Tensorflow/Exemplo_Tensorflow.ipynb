```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
```

#### Base de dados de frutas
0 - Maçã<br>
1 - Limão<br>
2 - Melão 

```python
# importando dataset
dataset = pd.read_csv("frutas.csv")
dataset.head()
```

```python
# extraindo características (x)
x = np.array(dataset.loc[:, ["ph", "peso(kg)", "diametro(cm)"]])

# extraindo resultado
d = np.array(dataset.loc[:,"fruta"])

# Embaralhando os dados
shuffle = np.random.permutation(len(x))
x = x[shuffle]
d = d[shuffle]

# Separando dados de treino (20%) e teste (20%)
limite = int(len(x) * 0.8)

x_treino = x[0:limite]
x_teste = x[limite: ]

d_treino = d[0:limite]
d_teste = d[limite: ]
```

### Criando modelo do keras

**Sequential** : Modelo sequencial, onde cada camada possui um vetor de entradas e um vetor de saídas<br>
**Flatten** : Camada de entrada, redimensiona os dados (caso a entrada seja uma matriz, por exemplo)<br>
**Dense** : Camada regular

Referências: https://www.tensorflow.org/tutorials/keras/classification?hl=pt-br , https://keras.io/about/

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(1, 3)),
    keras.layers.Dense(8, activation='sigmoid'),
    keras.layers.Dense(5, activation='sigmoid'),
    keras.layers.Dense(3, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_treino, d_treino, epochs=1000)
```

```python
previsao = model.predict(x_teste)
print(previsao)
```

```python
# Mapeando vetor para classe
def classificar(vetor):
    classes = []
    for y in vetor:
        maximo = np.argmax(y)
        classes.append(maximo)
        
    return classes
```

```python
classificar(previsao)
```

#### Comparando modelos

```python
modelo1 = keras.Sequential([
    keras.layers.Flatten(input_shape=(1, 3)),
    keras.layers.Dense(8, activation='sigmoid'),
    keras.layers.Dense(5, activation='sigmoid'),
    keras.layers.Dense(3, activation='softmax')
])

modelo2 = keras.Sequential([
    keras.layers.Flatten(input_shape=(1, 3)),
    keras.layers.Dense(5, activation='sigmoid'),
    keras.layers.Dense(4, activation='sigmoid'),
    keras.layers.Dense(3, activation='softmax')
])

modelo3 = keras.Sequential([
    keras.layers.Flatten(input_shape=(1, 3)),
    keras.layers.Dense(24, activation='sigmoid'),
    keras.layers.Dense(12, activation='sigmoid'),
    keras.layers.Dense(6, activation='sigmoid'),
    keras.layers.Dense(3, activation='softmax')
])

modelo1.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

modelo2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

modelo3.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
modelo1.fit(x_treino, d_treino, epochs=1000)
```

```python
modelo2.fit(x_treino, d_treino, epochs=1000)
```

```python
modelo3.fit(x_treino, d_treino, epochs=1000)
```

```python
loss1, acc1 = modelo1.evaluate(x_teste,  d_teste, verbose=2)
loss2, acc2 = modelo2.evaluate(x_teste,  d_teste, verbose=2)
loss3, acc3 = modelo3.evaluate(x_teste,  d_teste, verbose=2)
```

```python
print(f"Modelo 1:\nLoss: {loss1}\nAccuracy: {acc1}")
print(f"Modelo 2:\nLoss: {loss2}\nAccuracy: {acc2}")
print(f"Modelo 3:\nLoss: {loss3}\nAccuracy: {acc3}")
```
