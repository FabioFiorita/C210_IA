```python
class Utils:
    
    @staticmethod
    def euclidean_distance(p1, p2):
        '''
        Calcula a distância euclidiana
        '''
        position_1 = np.array(p1.position)
        position_2 = np.array(p2.position)
        distance = np.sqrt(sum((position_1 - position_2)**2))
            
        return distance
```
